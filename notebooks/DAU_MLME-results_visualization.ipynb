```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('seaborn')
```

```python
# freq models canola

r2_simple_c =  [0.3283,
                0.3960,
                0.3355,
                0.3288,
                0.4049,
                0.2748,
                0.3392]

r2_complex_c = [0.3408,
                0.4034,
                0.3400,
                0.3657,
                0.3441,
                0.2548,
                0.3289]

spearman_simple_c =  [0.5801,
                      0.6516,
                      0.6078,
                      0.5987,
                      0.6380,
                      0.5441,
                      0.6133]
                      
spearman_complex_c = [0.5888,
                      0.6489,
                      0.6055,
                      0.6239,
                      0.6045,
                      0.5477,
                      0.5808]
```

```python
# r2 canola freq
# freq models arabidopsis r2

fig = plt.figure(figsize=(8,5))
fig.patch.set_facecolor('white')

labels = ["100", "010", "110", "001", "101", "110", "111"]

ind = np.arange(len(labels))
width = 0.4       
plt.bar(ind, r2_simple_c, width, label="Simple model")
plt.bar(ind+width, r2_complex_c, width, label="Complex models")

plt.xticks(ind+width/2, labels)

plt.legend()

x1,x2,y1,y2 = plt.axis()  
# plt.axis((x1,x2,y1,0.12))
plt.axis((x1,x2,y1,y2))

plt.xlabel("Experimental Setting")
plt.ylabel("r2 Coefficient")
plt.title("Canola: Freq Model r2 Test Scores")
plt.show()
```

```python
# spearman canola freq
# freq models arabidopsis spearman

fig = plt.figure(figsize=(8,5))
fig.patch.set_facecolor('white')

labels = ["100", "010", "110", "001", "101", "110", "111"]

ind = np.arange(len(labels))
width = 0.4       
plt.bar(ind, spearman_simple_c, width, label="Simple model")
plt.bar(ind+width, spearman_complex_c, width, label="Complex models")

plt.xticks(ind+width/2, labels)

plt.legend()

x1,x2,y1,y2 = plt.axis()  
plt.axis((x1,x2,y1,0.75))

plt.xlabel("Experimental Setting")
plt.ylabel("Spearman Coefficient")
plt.title("Canola: Freq Model Spearman Test Scores")
plt.show()
```
